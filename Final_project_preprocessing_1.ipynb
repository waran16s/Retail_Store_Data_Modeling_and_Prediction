# **Preprocessing**

```python
#libaries
import pandas as pd
import numpy as np
```

**Collection of data**

```python
#Reading sales datasets
df_sal = pd.read_csv("https://raw.githubusercontent.com/waran16s/Retail_Store_Data_Modeling_and_Prediction/refs/heads/main/dataset/sales_data_set.csv")
df_sal
```

```python
#datasets Features
df_fea = pd.read_csv("https://raw.githubusercontent.com/waran16s/Retail_Store_Data_Modeling_and_Prediction/refs/heads/main/dataset/Features_data_set.csv")
df_fea
```

```python
df_sal.columns,df_fea.columns
```

***Merging datasets***

```python
df = pd.merge(df_sal,df_fea,on=['Store','Date','IsHoliday'],how='left')
```

```python
df
```

```python
df1 = df.copy()
```

```python
#Checking Dtypes
df1.info()
```

```python
#Changing date format
df1['date'] = pd.to_datetime(df1['Date'],format='%d/%m/%Y')
```

```python
df1.dtypes
```

```python
df1['Date'].unique()
```

**Checking Non Presented dates**

```python
missing_dates = df_fea[~df_fea['Date'].isin(df1['Date'])].reset_index(drop=True)
```

```python
missing_dates
```

```python
missing_dates['date'] = pd.to_datetime(missing_dates['Date'],format='%d/%m/%Y')
```

```python
df_new = df1[['Store','Dept']]
```

```python
df_new = df_new.drop_duplicates().reset_index(drop=True)
```

```python
df_new
```

```python
#checking nunique values
df_new['Store'].nunique(),df_new['Dept'].nunique()
```

```python
#combine data's
df_new1 = pd.merge(df_new,missing_dates,on=['Store'],how='outer')
```

```python
df_new1
```

```python
df_new1.drop('date',axis=1,inplace=True)
```

```python
df.shape,df_new1.shape
```

```python
df.columns
```

```python
df_new1.columns
```

```python
df_new2 = pd.concat([df,df_new1],ignore_index=True)
df_new2.shape
```

```python
df_new2
```

```python
#spliting dates data
df_new2['Date'] = pd.to_datetime(df_new2['Date'],format='%d/%m/%Y')

df_new2['Day'] = pd.to_datetime(df_new2['Date'],format='%d/%m/%Y').dt.day

df_new2['Month'] = pd.to_datetime(df_new2['Date'],format='%d/%m/%Y').dt.month

df_new2['Year'] = pd.to_datetime(df_new2['Date'],format='%d/%m/%Y').dt.year
```

```python
df_new2
```

**Combining store datasets**

```python
df_sto = pd.read_csv("https://raw.githubusercontent.com/waran16s/Retail_Store_Data_Modeling_and_Prediction/refs/heads/main/dataset/stores_data_set.csv")
df_sto
```

**Merging data**

```python
df = pd.merge(df_new2,df_sto,on=['Store'],how='left')
df
```

```python
#setting numerical value
df['IsHoliday'] = df['IsHoliday'].map({False:0,True:1})
df['Type'] = df['Type'].map({'A':1,'B':2,'C':3})
```

```python
df
```

```python
df = df.sort_values(by=['Date','Store']).reset_index(drop=True)
```

```python
df.head(10)
```

```python
df.to_csv('merged_datasets.csv',index=False)
```
